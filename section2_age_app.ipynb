# Age Application
## Requirements:
* Get age of user as input
* Print how many seconds the user has lived
* use input(), int(), and print() functions 
* use the format() string method

```python
age = int(input('Enter your age: '))
```

```python
print("You have lived for {} seconds.".format(age * 365 * 24 * 60 * 60))
```

# Age in seconds function

```python
def age_yr2sec(age):
    return age * 365 * 24 * 60 * 60
```

```python
print("You have lived for {} seconds.".format(age_yr2sec(age)))
```
